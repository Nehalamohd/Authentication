```python
# import library for preprocessing
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
import nltk
nltk.download('punkt_tab')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# preprocess the text
def preprocessing(text):
  text=text.lower()
  text=re.sub(r'[^A_za_z\w\s#]','',text)
  text_tokens = word_tokenize(text)
  filtered_text = [lemmatizer.lemmatize(w) for w in text_tokens if not w in stop_words]
  return " ".join(filtered_text)

```

```python
df.text=df['text'].apply(preprocessing)
```

```python
# use textblob
from textblob import TextBlob
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
df.loc[:, 'polarity']= df['text'].apply(polarity)
```

```python
df.head(5)
```

```python

def sentiment(label):
    if label <0:
        return "Negative"
    elif label ==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
df['sentiment'] = df['polarity'].apply(sentiment)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
def extract_named_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities
```

```python
df['entities'] = df['text'].apply(extract_named_entities)
```

```python
print(df[['entities']].head())
```

```python
# for converting words to numeric here use tfidf vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector=TfidfVectorizer()
```

```python
x=vector.fit_transform(df['text'])
```

```python

y=df['sentiment']
```

```python
#train the model
from sklearn.model_selection import train_test_split
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print("Size of x_train:", (x_train.shape))
print("Size of y_train:", (y_train.shape))
print("Size of x_test:", (x_test.shape))
print("Size of y_test:", (y_test.shape))
```

```python
#here first model is logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
logreg_pred = logreg.predict(x_test)
```

```python
logreg_acc = accuracy_score(logreg_pred, y_test)
logreg_acc
```

```python
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

```python
#to plot the confusion_matrix
cm = confusion_matrix(y_test, logreg_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=logreg.classes_)
disp.plot()
```

```python
# secoond model is SVC
```

```python
from sklearn.svm import LinearSVC
```

```python
SVCmodel = LinearSVC()
SVCmodel.fit(x_train, y_train)
```

```python
svc_pred = SVCmodel.predict(x_test)
svc_acc = accuracy_score(svc_pred, y_test)
svc_acc
```
